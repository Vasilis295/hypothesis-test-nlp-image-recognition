```python
df=pd.read_csv('wine.csv')
```

```python
sns.scatterplot(df['alcohol'],df['sulphates'],hue=df['quality'])
```

```python
df_high_a=df[df['alcohol']>10.5]
```

```python
df_low_a=df[df['alcohol']<=10.5]
```

```python
df_high_a1=df_high_a[['alcohol','quality']]
```

```python
df_low_a1=df_low_a[['alcohol','quality']]
```

```python
df_good=df[df['quality']=='good']
```

```python
df_bad=df[df['quality']=='bad']
```

```python
df_good_alcohol=df_good['alcohol']
```

```python
df_bad_alcohol=df_bad['alcohol']
```

```python
t_statistic, p_value = ttest_ind(df_good_alcohol,df_bad_alcohol)
print(t_statistic, p_value)
```

```python
p_value/2
```

```python
t_statistic, p_value = ttest_ind(df_high_a1['quality'],df_low_a1['quality'])
print(t_statistic, p_value)
```
